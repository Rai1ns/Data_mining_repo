```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from scipy.stats import f_oneway
warnings.filterwarnings('ignore')
```

```python
dengue = pd.read_csv('dataset\\dengue.csv')
```

<h1><span style = color:Red>Insight No:1</span> Dengue Cases on year 2020</h1>
<h1><span style = color:Red>Insight No:2</span> The Highest number of dengue cases in a month in 2020.</h1>
<h1><span style = color:Red>Insight No:3</span> The Lowest number of dengue cases in a month in 2020.</h1>

```python
year = dengue[dengue['Year']== 2020]
month = year.groupby('Month')['Dengue_Cases'].sum()

month.plot(kind='bar', color="Teal", edgecolor="black")
plt.gcf().set_size_inches(13,8)
plt.xlabel("Month")
plt.ylabel("Total Dengue Cases")
plt.title("Dengue Cases Per Month in 2020")
plt.xticks(rotation=45)
plt.show()
```

<h1><span style = color:red >Insight No:4</span> Number of Death Cause by Dengue in Every Year</h1>
<h1><span style = color:red >Insight No:5</span> Highest Number of Death Cause by Dengue in Every Year</h1>
<h1><span style = color:red >Insight No:6</span> Lowest Number of Death Cause by Dengue in Every Year</h1>

```python
year_deaths = dengue.groupby('Year')['Dengue_Deaths'].sum()
year_deaths.plot(kind='bar', color="Teal", edgecolor="black")
plt.gcf().set_size_inches(13,8)
plt.title('Dengue Deaths in Every Year')
plt.ylabel('Dengue Deaths')
plt.xlabel('Year')
plt.xticks(rotation=45)
plt.show()
```

```python
year_deaths = dengue.groupby('Year')['Dengue_Deaths'].sum()
plt.pie(year_deaths, labels = year_deaths.index, autopct = '%1.1f%%',colors=['teal','cyan','red','orange','purple'])
plt.gcf().set_size_inches(13,8)
plt.legend()
plt.show()
```

```python
every_region_cases = dengue.groupby('Region')['Dengue_Cases'].sum().reset_index()
top_5_highest_cases = every_region_cases.nlargest(5,'Dengue_Cases')

plt.barh(top_5_highest_cases['Region'], top_5_highest_cases['Dengue_Cases'],  color="Teal", edgecolor="black")
plt.title('Top 5 Region for Highest Cases of Dengue')
plt.xlabel('Dengue Cases')
plt.ylabel('Regions')
plt.gcf().set_size_inches(13,8)
plt.show()
```
