```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import scipy.stats
from scipy.stats import f_oneway
```

```python
accident = pd.read_csv('dataset\\accident_data.csv')
```

```python
accident
```

```python
accident.info()
```

<h1>DROPPING COLUMNS

```python
accident.drop(columns='Index', inplace=True)
```

```python
accident
```

<h1>CONVERTING OBJECT TO DATE TIME USING PANDAS</h1>

```python
accident['Accident Date'] = pd.to_datetime(accident['Accident Date'], dayfirst=True, errors = 'coerce')
```

```python
accident
```

<h1>DISECT DATE AND CREATE NEW COLUMNS</h1>

```python
accident['Month'] = accident['Accident Date'].dt.month
```

```python
accident['Day'] = accident['Accident Date'].dt.day
accident['Year'] = accident['Accident Date'].dt.year
accident['DayOfWeek'] = accident['Accident Date'].dt.dayofweek
```

```python
accident
```

<h1>TREND ANALYSIS

<h2>NUMBER OF ACCIDENTS PER YEAR

```python
accident_year = accident["Year"].value_counts()
```

```python
accident_year
```

```python
accident_year.plot(kind = "bar", color="red", edgecolor="black")
plt.title("ACCIDENTS PER YEAR")
plt.xlabel("YEAR")
plt.ylabel("NUMBER OF ACCIDENTS")
plt.gcf().set_size_inches(12,8)
plt.show()
```

<p> BASE ON THE FINDING WE SEE THAT THE YEAR WITH THE MOST NUMBER OF ACCIDENTS IS THE YEAR 2019

<h1>NUMBER OF ACCIDENTS PER MONTH

```python
accident_month = accident["Month"].value_counts().sort_index()
accident_month
```

```python
accident_month.plot(kind = "bar", color="red", edgecolor="black")
plt.title("ACCIDENTS PER MONTH")
plt.xlabel("MONTHS")
plt.ylabel("NUMBER OF ACCIDENTS")
plt.gcf().set_size_inches(12,8)
plt.show()
```

<p> BASE ON THE FINDING WE SEE THAT THE MONTH WITH THE MOST NUMBER OF ACCIDENTS IS THE MONTH OF NOVEMBER

<H1>NUMBER OF ACCIDENTS PER DAY OF WEEK

```python
accident_dow = accident["DayOfWeek"].value_counts().sort_index()
```

```python
accident_dow
```

```python
accident_dow.plot(kind = "bar", color="red", edgecolor="black")
plt.title("ACCIDENTS PER DAY OF A WEEK")
plt.xlabel("DAYOFWEEK")
plt.ylabel("NUMBER OF ACCIDENTS")
plt.gcf().set_size_inches(12,8)
plt.show()
```

<p> BASE ON THE FINDING WE SEE THAT THE DAY OF WEEK WITH THE MOST NUMBER OF ACCIDENTS IS THURSDAY

<H1>NUMBER OF ACCIDENTS PER MONTH UNDER LIGHT CONDITIONS

```python
accident_month_light = accident.groupby(['Month', "Light_Conditions"]).size()
```

```python
accident_month_light
```

```python
accident_month_light = accident.groupby(['Month', "Light_Conditions"]).size().unstack()
```

```python
accident_month_light
```

```python
accident_month_light.plot(kind = "bar", edgecolor="black", stacked=True)
plt.title("ACCIDENTS PER MONTH UNDER LIGHT CCONDITIONS")
plt.xlabel("MONTHS")
plt.ylabel("LIGHT CONDITIONS")
plt.gcf().set_size_inches(12,8)
plt.legend(title="LIGHT CONDITION", loc='upper left', bbox_to_anchor=(1.05, 1))
plt.show()
```

```python
accident_dayofweek = accident.groupby(['DayOfWeek', "Road_Surface_Conditions"]).size().unstack()
accident_dayofweek
```

<h1>NUMBER OF ACCIDENTS PER DAY OF WEEK UNDER ROAD SURFACE CONDITIONS

```python
accident_dayofweek.plot(kind = "bar", edgecolor="black", stacked=True)
plt.title("ACCIDENTS PER DAY OF WEEK UNDER ROAD SURFACE CONDITION")
plt.xlabel("DAY OF WEEK")
plt.ylabel("ROAD SURFACE CONDITIONS")
plt.gcf().set_size_inches(12,8)
plt.legend(title="ROAD SURFACE CONDITION", loc='upper left', bbox_to_anchor=(1.05, 1))
plt.show()
```
