```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from scipy.stats import f_oneway
warnings.filterwarnings('ignore')
import folium
from folium.plugins import HeatMap
```

```python
terrorist = pd.read_csv('dataset\\Global_terrorist.csv', encoding='ISO-8859-1')
```

```python
terrorist
```

```python
filter1 = terrorist[terrorist['region_txt'] == 'South America']
filter2 = filter1[filter1['iyear'] >= 2000]
filter3 = filter2[filter2['attacktype1_txt'] == 'Bombing/Explosion']
filter4 = filter3[filter3['nkill'] >= 10]
filter4
```

```python
spatial_d = filter4.groupby(['longitude', 'latitude']).size().reset_index(name='terror_count')
spatial_d
```

```python
region_assin = terrorist[(terrorist['region_txt'] == 'Southeast Asia') & (terrorist['attacktype1_txt'] == 'Assassination')]
year_kill = region_assin[(region_assin['iyear']>=2000) & (region_assin['nkill']>=5)] 

regional_spatial = year_kill.groupby(['latitude','longitude']).size().reset_index(name = 'Kill Count')

regional_heat_data = regional_spatial[['latitude','longitude','Kill Count']]

regional_base_map = folium.Map(location=[30,40],zoom_start = 4)

HeatMap(regional_heat_data).add_to(regional_base_map)

regional_base_map
```

```python
city_hijack = terrorist[(terrorist['region_txt'] == 'North America') &
            (terrorist['attacktype1_txt'] == 'Hijacking') &
            (terrorist['iyear'] >=1970)]

city_spatial = city_hijack.groupby(['longitude','latitude']).size().reset_index(name = 'Hijack Count')

city_heat_data = city_spatial[['latitude','longitude','Hijack Count']]

city_base_map = folium.Map(location=[30,40],zoom_start = 4)

HeatMap(city_heat_data).add_to(city_base_map)

city_base_map
```

```python
target_goverment = terrorist[(terrorist['region_txt'] == 'Eastern Europe') & 
            (terrorist['targtype1_txt'] == 'Government (Diplomatic)') & 
            (terrorist['iyear'] >=2005) &
            (terrorist['attacktype1_txt']== 'Bombing/Explosion')]

target_spatial = target_goverment.groupby(['longitude','latitude']).size().reset_index(name = 'Bombing Count')

target_heat_data = target_spatial[['latitude','longitude','Bombing Count']]

target_base_map = folium.Map(location=[30,40],zoom_start = 4)

HeatMap(target_heat_data).add_to(target_base_map)

target_base_map
```

```python
filtered_df = terrorist[(terrorist['region_txt'] == 'East Asia') & (terrorist['iyear'] >= 2008)]

# Aggregate incidents by location
aggregated_df = filtered_df.groupby(['latitude', 'longitude']).size().reset_index(name='incident_count')

# # Normalize weights (optional)
aggregated_df['normalized_weight'] = (aggregated_df['incident_count'] - aggregated_df['incident_count'].min()) / \
                                (aggregated_df['incident_count'].max() - aggregated_df['incident_count'].min())

# Prepare heat_data
heat_data = aggregated_df[['latitude', 'longitude', 'normalized_weight']].values.tolist()

# Create a base map
m = folium.Map(location=[30, 40], zoom_start=4)

# Add heatmap layer
HeatMap(heat_data).add_to(m)

# Save or display the map
m.save('terrorism_heatmap.html')
m
```

```python
crime_df = terrorist[(terrorist['nkill'] >= 10) & (terrorist['attacktype1_txt'] == 'Bombing/Explosion')]
aggregated_crime = crime_df.groupby(['latitude', 'longitude']).size().reset_index(name='incident_count')
aggregated_crime['normalized_weight'] = (aggregated_crime['incident_count'] - aggregated_crime['incident_count'].min()) / \
                                     (aggregated_crime['incident_count'].max() - aggregated_crime['incident_count'].min())
heat_data = aggregated_crime[['latitude', 'longitude', 'normalized_weight']].values.tolist()
crime_insight = folium.Map(location = [30, 40], zoom_start = 8)
HeatMap(heat_data).add_to(crime_insight)
crime_insight.save('terrorism_heatmap.html')
crime_insight
```
